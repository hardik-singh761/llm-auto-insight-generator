```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of Property Prices (in Lakhs)')
plt.xlabel('Price (Lakhs)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='area_type', y='price', data=df, palette='viridis')
plt.title('Average Price by Area Type')
plt.xlabel('Area Type')
plt.ylabel('Average Price (Lakhs)')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_sqft', y='price', hue='area_type', data=df, s=100)
plt.title('Price vs Total Square Feet')
plt.xlabel('Total Sqft')
plt.ylabel('Price (Lakhs)')
plt.show()
```
